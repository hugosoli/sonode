```python
from decimal import Decimal
from pprint import pprint
import boto3
import requests
import sys
import os
import subprocess
import os.path
from os import path
import json
```

```python
def update_audio(id, analysis, dynamodb=None):
    if not dynamodb:
        dynamodb = boto3.resource('dynamodb')

    table = dynamodb.Table('periferia')

    response = table.update_item(
        Key={
            'id': id
        },
        UpdateExpression="set analyzis=:a",
        ExpressionAttributeValues={
            ':a': analysis
        },
        ReturnValues="UPDATED_NEW"
    )
    return response
```

```python
def analizador(audio):
    print("analizando " + audio['id'])
    file_name, file_extension = os.path.splitext(audio['id'])
    if not(path.exists("../sonode-data/" + file_name + ".json")):
        subprocess.call(["essentia_streaming_extractor_freesound", "../sonode-data/" + audio['id'], "../sonode-data/" + file_name + ".json"])
    else:
        print(audio['id'] + " already analyzed")
```

```python
def descargador(audio):
    print(audio['id'])
    if not(path.exists("../sonode-data/" + audio['id'])):
        url = 'https://periferia.s3.amazonaws.com/' + audio['id']
        r = requests.get(url, allow_redirects=True)
        open("../sonode-data/" + audio['id'], 'wb').write(r.content)
    else:
        print(audio['id'] + " already downloaded")
```

```python
def update_audio(id, analysis, dynamodb=None):
    if not dynamodb:
        dynamodb = boto3.resource('dynamodb')

    table = dynamodb.Table('periferia')

    response = table.update_item(
        Key={
            'id': id
        },
        UpdateExpression="set analysis=:a",
        ExpressionAttributeValues={
            ':a': analysis
        },
        ReturnValues="UPDATED_NEW"
    )
    return response
```

```python
for audio in allAudios:
    descargador(audio)
```

```python
for audio in allAudios:
    analizador(audio)
```

```python
for audio in allAudios:
    file_name, file_extension = os.path.splitext(audio['id'])
    if path.exists("../sonode-data/" + file_name + ".json"):
        with open("../sonode-data/" + file_name + ".json") as f:
            data = json.load(f,parse_float=Decimal)
            update_audio(audio['id'],data)
```
